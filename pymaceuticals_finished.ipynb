## Observations and Insights 

1) Out of the treatments, Capomulin, Ramicane, Infubinol, and Ceftamin, Capomulin had the most consistent results, as illustrated by the lower IQR. The low IQR demonstrates consistency because there are more similar results in a smaller range

2) Tumor volume is moderately correlated (r=0.84) to mouse weight in Capomulin treated mice. As such, success of the treatment should be measured in percent decrease, rather than just the final volume, since the volume seems to, at least in part, be dependent on the mouse's weight. (Unless the mouse's weight is affected by tumor size too, depending on the density of the tumors)

3) Ramicane and Capomulin treated mice have lower average tumor volumes than the other treatments (~40mm3 vs >52mm3). At first glance at this data, it would seem those two treatments are possibly the most promising.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Remove duplicate rows by id and timepoint, keeping first instance of duplicates
# study_results = study_results.drop_duplicates(subset=["Mouse ID","Timepoint"])

# Combine the data into a single dataset
combined = pd.merge(mouse_metadata,study_results)
# Display the data table for preview
combined
```

```python
# Checking the number of mice.
len(combined["Mouse ID"].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
m_id = combined[combined.duplicated(['Mouse ID','Timepoint'],keep=False)]['Mouse ID'].unique()
m_id
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
extra_mouse_data = combined[combined['Mouse ID'] == m_id[0]]
extra_mouse_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_df = combined[combined['Mouse ID'] != m_id[0]]
cleaned_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(cleaned_df['Mouse ID'].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.
regimen = cleaned_df.groupby('Drug Regimen')['Tumor Volume (mm3)']
regimen_mean = regimen.mean()
regimen_median = regimen.median()
regimen_var = regimen.var()
regimen_std = regimen.std()
regimen_sem = regimen.sem()
dr_stats = pd.DataFrame({'Mean':regimen_mean,\
                         'Median':regimen_median,\
                         'Variance':regimen_var,\
                         'Standard Deviation':regimen_std,\
                         'SEM':regimen_sem})
dr_stats
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line
tumor_volume_stats = cleaned_df.groupby('Drug Regimen').agg({'Tumor Volume (mm3)':['mean','median','var','std','sem']})

tumor_volume_stats
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
tp_df = pd.DataFrame(cleaned_df.groupby('Mouse ID').max()['Timepoint'])

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
last_tp_df = pd.merge(tp_df,cleaned_df,on=['Mouse ID','Timepoint'])
last_tp_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin','Ramicane','Infubinol','Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    # Locate the rows which contain mice on each drug and get the tumor volumes

trtmnt_df = last_tp_df.loc[(last_tp_df['Drug Regimen'].isin(treatments)),
                           ['Mouse ID','Drug Regimen','Tumor Volume (mm3)']]
for drug in treatments:
    #Draw Boxplots
    plt.boxplot(trtmnt_df.loc[trtmnt_df['Drug Regimen'] == drug,'Tumor Volume (mm3)'])
    plt.title(f'Tumor Volume (mm3) in Mice Treated with {drug}')
    plt.show()
    
    #Find quartile info
    quartiles = trtmnt_df.loc[trtmnt_df['Drug Regimen'] == drug,'Tumor Volume (mm3)'].quantile([0.25,0.5,0.75])
    lowerq = quartiles[.25]
    upperq = quartiles[.75]
    iqr = upperq-lowerq

    # Determine outliers using upper and lower bounds
    upper_bound = upperq + (1.5*iqr)
    lower_bound = lowerq - (1.5*iqr)

# Print out relevant information
    print(f"Mice Treated with {drug}:")
    print(f"The lower quartile of tumor volume is: {round(lowerq,2)}")
    print(f"The upper quartile of tumor volume is: {round(upperq,2)}")
    print(f"The interquartile range of tumor volume is: {round(iqr,2)}")
    print(f"The the median of tumor volume is: {quartiles[0.5]} ")
    print(f"Values below {round(lower_bound,2)} could be outliers.")
    print(f"Values above {round(upper_bound,2)} could be outliers.")
    print('--------------------------')
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin
# Filter out df for only capomulin and print out available options
capo_df = cleaned_df.loc[cleaned_df['Drug Regimen'] == 'Capomulin']
print(f"Valid ID Options: {capo_df['Mouse ID'].unique()}")

# get which mouse to track and filter out info for only that mouse
mouse_of_interest = input("Type the ID of the mouse you want to track: ")
interested_df = capo_df.loc[capo_df['Mouse ID'] == mouse_of_interest]

# plot info
plt.plot(interested_df['Timepoint'],interested_df['Tumor Volume (mm3)'])
plt.title(f"Tumor Progress for Mouse {mouse_of_interest}")
plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume (mm3)')
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
capo_avg = capo_df.groupby('Mouse ID').mean()

plt.scatter(capo_avg['Tumor Volume (mm3)'],capo_avg['Weight (g)'])
plt.title('Average Tumor Volume vs Mouse Weight')
plt.xlabel('Tumor Volume (mm3)')
plt.ylabel('Mouse Weight (g)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
corr = st.pearsonr(capo_avg['Tumor Volume (mm3)'],capo_avg['Weight (g)'])
(slope,intercept,r,p,stderr) = st.linregress(capo_avg['Tumor Volume (mm3)'],capo_avg['Weight (g)'])

print("Average Tumor Volume vs Mouse Weight:")
print(f"The correlation coefficient is: {round(corr[0],2)}")
# ^or print(f"The correlation coefficient is: {round(r,2)}")
print(f"The linear regression model can be expressed as: y = {round(slope,2)}x + {round(intercept,2)}")
```
